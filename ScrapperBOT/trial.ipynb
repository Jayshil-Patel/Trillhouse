```python
%pip install pyautogen>=0.2.3
from apikey import api
API_key = api
from typing import Dict, Union

from IPython import get_ipython
from IPython.display import Image

import autogen

config_list = [
    {
        'model': 'gpt-3.5-turbo-16k-0613',
        'api_key': API_key,
    }
]
```

```python
llm_config={
    "seed": 50,                     # for caching and reproducibility
    "config_list": config_list,     # which models to use
    "temperature": 0,               # for sampling
}

# create an AssistantAgent instance named "assistant"
agent_assistant = autogen.AssistantAgent(
    name="agent_assistant",
    llm_config=llm_config,
)

# create a UserProxyAgent that acts on the humans behalf
agent_proxy = autogen.UserProxyAgent(
    name="agent_proxy",
    human_input_mode="NEVER",           # NEVER, TERMINATE, or ALWAYS
                                            # TERMINATE - human input needed when assistant sends TERMINATE
    max_consecutive_auto_reply=10,
    is_termination_msg=lambda x: x.get("content", "").rstrip().endswith("TERMINATE"),
    code_execution_config={
        "work_dir": "New_iteration",     # path for file output of program
        "use_docker": False,            # True or image name like "python:3" to use docker image
    },
    llm_config=llm_config,
    system_message="""Reply TERMINATE if the task has been solved at full satisfaction.
                      Otherwise, reply CONTINUE, or the reason why the task is not solved yet."""
)

agent_proxy.initiate_chat(
    agent_assistant,
    message=task,
)
```

```python
import os

# Get the current directory of the 'trial.ipynb' file
current_directory = os.getcwd()

# Construct the path to the 'dom.html' file
file_path = os.path.join(current_directory, 'New_iteration', 'dom.html')

# Read the contents of the file with the specified encoding
with open(file_path, 'r', encoding='utf-8') as file:
    html_content = file.read()

# Now you can use the html_content variable which contains the content of 'dom.html'
```

```python


from langchain.text_splitter import CharacterTextSplitter
from langchain.embeddings.openai import OpenAIEmbeddings
from langchain.vectorstores import FAISS
from langchain.chains.question_answering import load_qa_chain
from langchain.llms import OpenAI
from langchain.callbacks import get_openai_callback
```

```python
 # split into chunks
text_splitter = CharacterTextSplitter(
        separator="\n",
        chunk_size=1000,
        chunk_overlap=200,
        length_function=len
      )
chunks = text_splitter.split_text(html_content)
      
      # create embeddings
embeddings = OpenAIEmbeddings(api_key = api)
knowledge_base = FAISS.from_texts(chunks, embeddings)
```
